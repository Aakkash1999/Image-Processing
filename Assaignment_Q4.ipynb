```python
#Question[04]

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

f = cv.imread(r'C:\Users\ALPHAXANDER\Downloads\shells.png', cv.IMREAD_GRAYSCALE)
assert f is not None

hist_f = cv.calcHist([f], [0], None, [256], [0,256])
g = cv.equalizeHist(f)

hist_g = cv.calcHist([g], [0], None, [256], [0,256])

fig, ax = plt.subplots(1,2,figsize=(16,8))
ax[0].plot(hist_f)
ax[0].set_title('Original Image - Histogram')
ax[1].plot(hist_g)
ax[1].set_title('Equalized Image - Histogram')
plt.show()



fig, ax = plt.subplots(1,2,figsize=(16,8))
f_rgb = cv.cvtColor(f, cv.COLOR_BGR2RGB)
g_rgb = cv.cvtColor(g, cv.COLOR_BGR2RGB)
ax[0].imshow(f_rgb)
ax[0].set_title('Original Image')
ax[1].imshow(g_rgb)
ax[1].set_title("Equalized Image")
plt.show()
```
