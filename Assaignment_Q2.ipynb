```python
#Question [02]

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

f = cv.imread(r'C:\Users\ALPHAXANDER\Downloads\brain_proton_density_slice.png')
assert f is not None

hist = cv.calcHist([f],[0],None,[256],[0,256])
plt.plot(hist)

#Gray matter

t1 = np.linspace(0, 0, 140)
t2 = np.linspace(0,250,0)
t3 = np.linspace(250, 256, 35)
t4 = np.linspace(256,0,0)
t5 = np.linspace(0,0,81)

t_1 = np.concatenate((t1, t2, t3, t4, t5), axis = 0).astype(np.uint8)
assert len(t_1) == 256

g = cv.LUT(f, t_1)
im = cv.cvtColor(g, cv.COLOR_BGR2RGB)

fig, ax = plt.subplots(1,3,figsize=(16,8))

ax[0].plot(t_1)
ax[0].set_title('Intensity Transformation')
ax[0].set_xlabel('Input intensity')
ax[0].set_ylabel('Output intensity')
ax[0].set_aspect('equal')
ax[1].imshow(f)
ax[1].set_title('The Original image')
ax[2].imshow(g)
ax[2].set_title('Gray Matter')

#White matter

t1 = np.linspace(0, 10, 150) 
t2 = np.linspace(10, 230 ,66)
t3 = np.linspace(230,255,40)

t_2 = np.concatenate((t1, t2, t3), axis = 0).astype(np.uint8)
assert len(t_2) == 256

g = cv.LUT(f, t_2)
im = cv.cvtColor(g, cv.COLOR_BGR2RGB)

fig, ax = plt.subplots(1,3,figsize=(16,8))

ax[0].plot(t_2)
ax[0].set_title('Intensity Transformation')
ax[0].set_xlabel('Input intensity')
ax[0].set_ylabel('Output intensity')
ax[0].set_aspect('equal')
ax[1].imshow(f)
ax[1].set_title('The Original image')
ax[2].imshow(g)
ax[2].set_title('White Matter')
```
